Import needed libaries and load cleaned data

```python
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd

df = pd.read_csv("marketing_campaign_data_cleaned.csv")
df
```

Plot clicks agains impressions to show linearity between them. More impressions turn into more clicks.

```python
plt.scatter(df["impressions"], df["total_clicks"])
plt.xlabel("Impressions", fontsize=14)
plt.ylabel("Total Clicks", fontsize=14)  
plt.show()
```

```python
pie_info = [
    (df["channel"] == "TikTok").sum(),
    (df["channel"] == "Facebook").sum(),
    (df["channel"] == "Email").sum(),
    (df["channel"] == "Instagram").sum(),
    (df["channel"] == "Google").sum(),
    (df["channel"] == "Google Ads").sum()
]
names = ["TikTok", "Facebook", "Email", "Instagram", "Google", "Google Ads"]

plt.pie(pie_info, labels=names, autopct='%1.1f%%')
plt.title("Traffic by platform")
plt.show()
```
